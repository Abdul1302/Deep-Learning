```python
import pandas as pd 
import numpy as np
```

# Generate Random Classification Dataset

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=100,n_features=2,n_classes=2,n_informative=1,n_redundant=0,
                        n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

# Plot the Points On Random Data

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

# Make Perceptron Function

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(x,y):
    x=np.insert(x,0,1,axis=1)  #add 1 for baisness
    weights=np.ones(x.shape[1]) #set the shape "weights" as shape of "x"
    lr=0.1  #learning rate
    
    for i in range(1000):  #loop work as epochs
        j=np.random.randint(0,100)  #select random int between "0" to "100" (signify random value selection from data)
        y_hat=step(np.dot(x[j],weights)) #if output>0 step function is "1" else "0"
        weights=weights+lr*(y[j]-y_hat)*x[j] #prediction of perceptron as the formula
        
    
    return weights[0],weights[1:] #weights[0]=intercept ,weights[1:]=Coefficient  
```

```python
intercept,coeff=perceptron(x,y)
```

```python
print(intercept)
print(coeff)
```

```python
#y=mx+b eq:Ax+By+c=0
m=-(coeff[0]/coeff[1]) 
b=-(intercept/coeff[1])
```

# Plot The Line

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
